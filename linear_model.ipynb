```python
from sklearn.base import BaseEstimator
import numpy as np
```

```python
class Linear_Model(BaseEstimator):
    """Linear model that works with linear data for predictions"""
    def __init__(self):
        self.theta = None

    def fit(self, X_train, y_train):
        X_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]
        X_trnsp = X_b.T
        self.theta = np.linalg.inv(X_trnsp @ X_b) @ X_trnsp @ y_train
        return self.theta
    
    def predict(self, X_test):
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        scores = X_test_b @ self.theta
        return scores
```
