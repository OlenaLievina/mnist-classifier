```python
from sklearn.base import BaseEstimator
import numpy as np
```

```python
class Stochatic_Linear_Model(BaseEstimator):
    def __init__(self):
        self.theta = None 

    def fit(self, X_train, y_train, n_iter=1000, eta=0.1):
        m, n = X_train.shape
        self.theta = np.random.randn(n+1, 1)
        
        data = np.c_[X_train, y_train]
        for i in range(n_iter):
            shuffled_arr = np.random.permutation(data)
            
            for sample in shuffled_arr:

                xi = sample[:n].reshape(1, n)
                yi = sample[n:].reshape(1, 1)

                sample_b = np.c_[1, xi]
                gradients = 2 * sample_b.T @ (sample_b @ self.theta - yi)
                self.theta = self.theta - eta * gradients

        return self.theta
    
    def predict(self, X_test):
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        scores = X_test_b @ self.theta
        return scores
```
