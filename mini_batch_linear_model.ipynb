```python
from sklearn.base import BaseEstimator
import numpy as np
```

```python
class Mini_Batch_Linear_Model(BaseEstimator):
    def __init__(self):
        self.theta = None
    def fit(self, X_train, y_train, eta=0.1, n_iter=1000, batch_size=32):
        m, n = X_train.shape
        self.theta = np.random.randn(n+1, 1)
        X_b = np.c_[np.ones((m, 1)), X_train]

        for epoch in range(n_iter):
            indices = np.random.permutation(m)
            X_b_sh = X_b[indices]
            y_sh = y_train[indices]

            for i in range(0, m, batch_size):
                xi = X_b_sh[i : i+batch_size]
                yi = y_sh[i : i+batch_size]

                gradients = 2/len(xi) *xi.T @ (xi @ self.theta - yi)
                self.theta = self.theta - eta * gradients

    def predict(self, X_test):
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        scores = X_test_b @ self.theta
        return scores
```
