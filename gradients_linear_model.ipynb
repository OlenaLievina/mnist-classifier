```python
from sklearn.base import BaseEstimator
import numpy as np
```

```python
class Gradients_Linear_Model(BaseEstimator):
    def __init__(self):
        self.thetas = None

    def fit(self, X_train, y_train, n_iter=1000, eta=0.1):
        self.thetas = np.random.randn(2, 1)
        X_bias = np.c_[np.ones((X_train.shape[0], 1)), X_train]
        for iteration in range(n_iter):
            gradients = 2/X_train.shape[0] *  X_bias.T @ (X_bias @ self.thetas - y_train)
            self.thetas = self.thetas - eta * gradients
        return self.thetas
    
    def predict(self, X_test):
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        scores = X_test_b @ self.thetas
        return scores
```
