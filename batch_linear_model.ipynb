```python
from sklearn.base import BaseEstimator
import numpy as np
```

```python
class Gradients_Linear_Model(BaseEstimator):
    def __init__(self):
        self.theta = None

    def fit(self, X_train, y_train, n_iter=1000, eta=0.1):
        m, n = X_train.shape
        self.theta = np.random.randn(n + 1, 1)
        X_bias = np.c_[np.ones((X_train.shape[0], 1)), X_train]
        for iteration in range(n_iter):
            gradients = 2/m * X_bias.T @ (X_bias @ self.theta - y_train)
            self.theta = self.theta - eta * gradients
        return self.theta
    
    def predict(self, X_test):
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        scores = X_test_b @ self.theta
        return scores
```
